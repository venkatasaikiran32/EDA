```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
#reading dataset
df=pd.read_csv('amazon.csv')
```

```python
df.head()
```

```python
#no need of product_id

df.drop('product_id',axis=1,inplace=True)
```

```python
df.columns
```

```python
df.info()
```

```python
for i in range(len(df.columns)):
  if  df.iloc[:,i].nunique() < 10 and df.iloc[:,i].dtype=='object':
    df.iloc[:,i]=df.iloc[:,i].astype('category')
```

```python
#as rating column only contains floating values we will convert it to float data type
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
```

```python
#rating count has discrete values
#so we can also convert it into numeric
df['review_content']=pd.to_numeric(df['review_content'],errors='coerce')
```

```python
df.head()
```

```python
#discounted_price	,actual_price	,discount_percentage
#we have to modify these columns
#in discounted_price we have rupees symbol we dont want it
df['discounted_price']=df['discounted_price'].apply(lambda x : x[1:])
df['actual_price']	=df['actual_price'].apply(lambda x:x[1:])
df['discount_percentage']=df['discount_percentage'].apply(lambda x:x[:-1])
```

```python
df['discount_percentage']=pd.to_numeric(df['discount_percentage'],errors='coerce')
df['discounted_price']=pd.to_numeric(df['discounted_price'],errors='coerce')
df['actual_price']=pd.to_numeric(df['actual_price'],errors='coerce')
```

```python
df.info()
```

6. what are the most popular product keywords?

```python
from collections import Counter
import re

# Extract keywords from product names
product_names = df['product_name'].dropna().tolist()
keywords = ' '.join(product_names).lower()
words = re.findall(r'\w+', keywords)
keyword_counts = Counter(words)

# Convert to DataFrame for easier viewing
keywords_df = pd.DataFrame(keyword_counts.items(), columns=['Keyword', 'Count']).sort_values('Count', ascending=False)
print("Most Popular Product Keywords:")
print(keywords_df.head(10))
```

7.For the top 5 most popular artists, what are their most popular tracks? List the track name for each artist

```python
#firstly top 5 most popular artists
df.groupby('Artist')['Popularity'].mean().sort_values(ascending=False).head(5)



```

```python
# Get the top 5 most popular artists
top_artists = df.groupby('Artist')['Popularity'].mean().nlargest(5).index

# Filter tracks of these artists and get the most popular tracks
top_tracks = df[df['Artist'].isin(top_artists)].sort_values(by='Popularity', ascending=False)
for artist in top_artists:
    artist_tracks = top_tracks[top_tracks['Artist'] == artist]
    print(f"\nMost Popular Tracks for {artist}:")
    print(artist_tracks[['Track Name', 'Popularity']].head(1))
```
